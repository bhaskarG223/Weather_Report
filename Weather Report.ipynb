1. Add Month and Season columns for grouping

```python
weather['Month'] = weather['Date/Time'].dt.month
weather['Season'] = weather['Month'] % 12 // 3 + 1
weather['Season'] = weather['Season'].replace({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})
```

```python
weather.head()
```

2.Extreme Events: Highest/Lowest Temperature

```python
extreme_events = weather[['Temp_C', 'Date/Time']].sort_values(by='Temp_C', ascending=False)
highest_temp = extreme_events.head(1)
lowest_temp = extreme_events.tail(1)
```
